```python
import streamlit as st
import pandas as pd

# Load your CSV file (replace 'your_csv_url' with the actual URL)
url = "notebooks/test-scripts-sample.csv"
df = pd.read_csv(url)

# Sidebar with genre selection
selected_genre = st.sidebar.selectbox("Select Genre:", ["All"] + df['genre'].unique().tolist())

# Filter the DataFrame based on the selected genre
if selected_genre != "All":
    df_filtered = df[df['genre'] == selected_genre]
else:
    df_filtered = df  # Display all rows if "All" is selected

# Main content area
st.title("Movie Genre Selection App")

# Display the selected genre
st.write(f"Showing movies in the genre: {selected_genre}")

# Display the filtered DataFrame
st.dataframe(df_filtered)
```
